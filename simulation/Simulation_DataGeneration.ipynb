```python
import numpy as np
import pandas as pd
import random

# Set the seed
random.seed(42)
# Parameters
n_firms = 5000
n_days = 365 * 40
r_prob_1 = 0.25  # Probability of drawing r from the half-normal distribution
r_prob_2 = 0.1  # Probability of drawing r from the half-normal distribution

# Create an empty list to store data for all firms
all_firms_data = []

for firm_id in range(1, n_firms + 1):
    # Generate e values from a normal distribution
    # Generate a random number between 0 and 1
    random_number = random.random()
    print(random_number)
    e_values = np.random.normal(0, 5*random_number, n_days)
    
    # Generate r values
    r_values = np.zeros(n_days)  # Start with all r values as 0
    r_patent_dummy = np.zeros(n_days)  # Start with all r values as 0
    r_science=np.zeros(n_days)
    r_single_dist=  np.zeros(n_days) 
    
    sigma_values = random_number
    
    r_draw = np.random.rand(n_days)  # Random numbers to decide when to draw from half-normal
    r_values[r_draw < r_prob_1] = np.abs(np.random.normal(0, random_number, sum(r_draw < r_prob_1)))
    r_patent_dummy[r_draw < r_prob_1]=1
    
    r_single_dist[r_draw < r_prob_1] = np.abs(np.random.normal(0, random_number, sum(r_draw < r_prob_1)))
    
    
    r_values[r_draw < r_prob_2] = np.abs(np.random.normal(0, random_number*1.1, sum(r_draw < r_prob_2)))
    r_science[r_draw < r_prob_2]=1
    
    
    
    # Store the data in a DataFrame for this firm
    firm_data = pd.DataFrame({
        'firm_id': firm_id,
        'day': np.arange(1, n_days + 1),
        'e': e_values,
        'x': r_values,
        'x_single_dist':r_single_dist,
        'patent_dummy':r_patent_dummy,
        'science_dummy':r_science,
        'noise_variance':sigma_values*5,
        'signal_variance_science':sigma_values*1.1,
        'signal_variance_noscience':sigma_values
    })
    
    all_firms_data.append(firm_data)

# Combine data from all firms
df = pd.concat(all_firms_data, ignore_index=True)


df.to_stata(r'C:\Users\jayho\OneDrive - Duke University\ABF_PatValue_Research\Code\simulatation\simulate_data_highpatdays.dta')
```
